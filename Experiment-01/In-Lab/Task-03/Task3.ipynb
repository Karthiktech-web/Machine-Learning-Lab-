### Objective

To visualize the Iris dataset using univariate and multivariate analysis techniques for better understanding of data distribution and relationships between features.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("iris.csv")

iris.head()
```

## Step 3

### Histogram

A histogram shows the distribution of a single numerical variable.

```python
plt.figure(figsize=(6,4))
plt.hist(iris["sepal_length"], bins=10)
plt.title("Histogram of Sepal Length 25eu02904")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

## Step 4

### Box Plot

A box plot displays the spread of data and helps identify outliers.

```python
plt.figure(figsize=(6,4))
plt.boxplot(iris["petal_length"])
plt.title("Box Plot of Petal Length 25eu02904")
plt.ylabel("Petal Length")
plt.show()
```

## Step 5

### Scatter Plot

A scatter plot shows the relationship between two numerical variables.

```python
plt.figure(figsize=(6,4))
plt.scatter(iris["sepal_length"], iris["petal_length"])
plt.title("Sepal Length vs Petal Length 25EU02904")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

# Conclusion

Successfully performed univariate and multivariate analysis on the Iris dataset using Matplotlib.

The histogram showed the distribution of sepal length, the box plot highlighted the spread of petal length, and the scatter plot illustrated the relationship between sepal length and petal length.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
iris = pd.read_csv("iris.csv")

# Histogram
plt.figure(figsize=(6,4))
plt.hist(iris["sepal_length"], bins=10)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()

# Box Plot
plt.figure(figsize=(6,4))
plt.boxplot(iris["petal_length"])
plt.title("Box Plot of Petal Length")
plt.ylabel("Petal Length")
plt.show()

# Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(iris["sepal_length"], iris["petal_length"])
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()

print("Task 3 Completed Successfully")
```
