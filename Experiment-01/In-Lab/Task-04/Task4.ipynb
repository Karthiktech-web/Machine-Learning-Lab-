### Objective

To visualize the Iris dataset using different types of graphs such as line chart, bar chart, pie chart, and histogram to better understand the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("iris.csv")

iris.head()
```

```python
plt.figure(figsize=(6,4))
plt.plot(iris["sepal_length"])
plt.title("Line Chart of Sepal Length 25EU02904")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
species_count = iris["species"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(species_count.index, species_count.values)
plt.title("Species Count 25EU02904")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(species_count.values,
        labels=species_count.index,
        autopct="%1.1f%%")

plt.title("Species Distribution 25eu02904")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(iris["petal_width"], bins=10)
plt.title("Histogram of Petal Width 25eu02904")
plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.show()
```

# Conclusion

Successfully visualized the Iris dataset using different types of graphs.

The line chart showed the trend of sepal length, the bar chart and pie chart displayed the distribution of flower species, and the histogram illustrated the frequency distribution of petal width.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

iris = pd.read_csv("iris.csv")

# Line Chart
plt.figure(figsize=(6,4))
plt.plot(iris["sepal_length"])
plt.title("Line Chart of Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()

# Bar Chart
species_count = iris["species"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(species_count.index, species_count.values)
plt.title("Species Count")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# Pie Chart
plt.figure(figsize=(6,6))
plt.pie(species_count.values,
        labels=species_count.index,
        autopct="%1.1f%%")
plt.title("Species Distribution")
plt.show()

# Histogram
plt.figure(figsize=(6,4))
plt.hist(iris["petal_width"], bins=10)
plt.title("Histogram of Petal Width")
plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.show()

print("Task 4 Completed Successfully")
```
