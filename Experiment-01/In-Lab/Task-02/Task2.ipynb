### Objective

## Step 1

### Import Required Libraries

NumPy is used for numerical computations.

Pandas is used for data analysis and data manipulation.

```python
import numpy as np
import pandas as pd
```

## Step 2

### Loading the Iris Dataset

The Iris dataset is loaded into a Pandas DataFrame using the `read_csv()` function.

```python
iris = pd.read_csv("iris.csv")

iris.head()
```

## Step 3

# Dataset Shape

The `shape` attribute returns the number of rows and columns in the dataset.

It helps us understand the size of the dataset before performing analysis.

```python
print("Shape of the dataset:", iris.shape)
```

## Step 4

# Data Types of Each Column

The `dtypes` attribute displays the data type of each column in the dataset.

Knowing the data type helps determine which statistical operations can be performed.

```python
iris.dtypes
```

## Step 5

# Descriptive Statistical Summary

The `describe()` function generates summary statistics for all numerical columns.

It provides information such as:
- Count
- Mean
- Standard Deviation
- Minimum
- Maximum
- Quartiles (25%, 50%, 75%)

```python
iris.describe()
```

## Step 6

# Mean (Average)

The **mean** is the average value of a numerical column.

It is calculated by adding all values in a column and dividing by the total number of values.

The mean helps us understand the central tendency of the dataset.

```python
iris.mean(numeric_only=True)
```

## Step 7

# Median

The **median** is the middle value of a dataset after arranging the values in ascending order.

If the number of observations is even, the median is the average of the two middle values.

The median is less affected by outliers than the mean.

```python
iris.median(numeric_only=True)
```

## Step 8

# Mode

The **mode** is the value that occurs most frequently in a dataset.

A dataset may have one mode, multiple modes, or no mode.

```python
iris.mode()
```

## Step 9

# Variance

Variance measures how far the data values are spread from the mean.

A larger variance indicates greater variability in the data.

```python
iris.var(numeric_only=True)
```

## Step 10

# Standard Deviation

Standard deviation is the square root of the variance.

It indicates how much the values deviate from the mean.

A low standard deviation means the values are close to the mean, while a high standard deviation indicates greater variation.

```python
iris.std(numeric_only=True)
```

## Step 11

# Correlation Matrix

Correlation measures the relationship between two numerical variables.

The correlation coefficient ranges from **-1 to +1**.

- **+1** → Perfect positive correlation
- **0** → No correlation
- **-1** → Perfect negative correlation

The correlation matrix helps identify relationships between different features.

```python
iris.corr(numeric_only=True)
```

## Step 12

# Checking Missing Values

Missing values can affect the accuracy of data analysis and machine learning models.

The `isnull().sum()` function counts the number of missing values in each column.

```python
iris.isnull().sum()
```

## Step 13

# Finding Unique Values

The `nunique()` function returns the number of unique values in each column.

It helps understand the diversity of values present in the dataset.

```python
iris.nunique()
```

## Step 14

# Counting Species

The `value_counts()` function counts the occurrences of each category in the dataset.

It is useful for understanding the distribution of categorical variables.

```python
iris["species"].value_counts()
```

```python
# =====================================================
# Experiment 01 - Task 02
# Descriptive Statistical Analysis on Iris Dataset
# =====================================================

# Import Libraries
import numpy as np
import pandas as pd

# Load Dataset
iris = pd.read_csv("iris.csv")

# Display First Five Rows
print("========== First Five Rows ==========")
print(iris.head())

# Display Last Five Rows
print("\n========== Last Five Rows ==========")
print(iris.tail())

# Dataset Shape
print("\n========== Dataset Shape ==========")
print("Rows and Columns:", iris.shape)

# Data Types
print("\n========== Data Types ==========")
print(iris.dtypes)

# Column Names
print("\n========== Column Names ==========")
print(iris.columns)

# Statistical Summary
print("\n========== Descriptive Statistics ==========")
print(iris.describe())

# Mean
print("\n========== Mean ==========")
print(iris.mean(numeric_only=True))

# Median
print("\n========== Median ==========")
print(iris.median(numeric_only=True))

# Mode
print("\n========== Mode ==========")
print(iris.mode())

# Variance
print("\n========== Variance ==========")
print(iris.var(numeric_only=True))

# Standard Deviation
print("\n========== Standard Deviation ==========")
print(iris.std(numeric_only=True))

# Correlation Matrix
print("\n========== Correlation Matrix ==========")
print(iris.corr(numeric_only=True))

# Missing Values
print("\n========== Missing Values ==========")
print(iris.isnull().sum())

# Number of Unique Values
print("\n========== Unique Values ==========")
print(iris.nunique())

# Species Count
print("\n========== Species Count ==========")
print(iris["species"].value_counts())

print("\n========== Task 2 Completed Successfully ==========")
```
