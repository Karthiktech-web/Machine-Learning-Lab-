## Step 1

### Import Required Libraries

NumPy is used for numerical computations.

Pandas is used for loading and analyzing datasets.

```python
import numpy as np
import pandas as pd
```

## Step 2

# Creating a NumPy Array

NumPy (Numerical Python) is a Python library used for numerical computations.

A NumPy array stores multiple values of the same data type and allows mathematical operations to be performed efficiently.

```python
# Creating a NumPy array

numbers = np.array([10, 20, 30, 40, 50])

print(numbers)
```

## Step 3

# Basic Operations on NumPy Arrays

NumPy provides built-in functions to perform statistical and mathematical operations on arrays.

```python
print("Array :", numbers)

print("Maximum Value :", numbers.max())

print("Minimum Value :", numbers.min())

print("Sum :", numbers.sum())

print("Mean :", numbers.mean())
```

## Step 4

# Introduction to Pandas DataFrame

A DataFrame is a two-dimensional data structure provided by the Pandas library.

It stores data in rows and columns, similar to an Excel spreadsheet or a database table.

DataFrames make it easy to organize, analyze, and manipulate datasets.

```python
# Creating a dictionary

student_data = {
    "Name": ["Karthik", "Rahul", "Priya", "Anjali"],
    "Age": [20, 21, 20, 22],
    "Branch": ["AIML", "CSE", "ECE", "IT"],
    "Marks": [91, 85, 88, 95]
}

# Convert dictionary into DataFrame

df = pd.DataFrame(student_data)

# Display the DataFrame

df
```

## Step 5

# Basic DataFrame Operations

Pandas provides several functions to understand the structure and contents of a dataset.

```python
# Display first 2 rows

print(df.head(2))

print()

# Display last 2 rows

print(df.tail(2))

print()

# Shape of the DataFrame

print("Shape :", df.shape)

print()

# Column names

print("Columns :", df.columns)
```

## Step 6

# Loading the Iris Dataset

The Iris dataset is one of the most popular datasets used in Machine Learning.

It contains measurements of iris flowers belonging to three different species.

The dataset contains:

- Sepal Length
- Sepal Width
- Petal Length
- Petal Width
- Species

```python
# Load the Iris dataset

iris = pd.read_csv("iris.csv")

# Display the first five rows

iris.head()
```

## Step 8

# Displaying the Last Five Rows

The `tail()` function displays the last five rows of the dataset. It helps verify the ending records and inspect the dataset.


```python
iris.tail()
```

## Step 9

# Finding the Shape of the Dataset

The `shape` attribute returns the number of rows and columns present in the dataset.

```python
print("Shape of the dataset:", iris.shape)
```

## Step 10

# Displaying Column Names

The `columns` attribute displays the names of all columns in the dataset.

```python
print(iris.columns)
```

## Step 11

# Dataset Information

The `info()` function provides information about the dataset, including the number of rows, columns, data types, and missing values.

```python
iris.info()
```

## Step 12

# Statistical Summary

The `describe()` function generates descriptive statistics for numerical columns in the dataset.

```python
iris.describe()
```

## Step 13

# Checking Missing Values

The `isnull().sum()` function checks for missing values in each column of the dataset.

```python
iris.isnull().sum()
```
