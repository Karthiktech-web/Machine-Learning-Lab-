```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic Dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)

print("First Five Rows")
print(titanic.head())

print("\nShape")
print(titanic.shape)

print("\nData Types")
print(titanic.dtypes)

print("\nStatistical Summary")
print(titanic.describe())

print("\nMissing Values")
print(titanic.isnull().sum())

print("\nSurvival Count")
print(titanic["Survived"].value_counts())

print("\nPassenger Class")
print(titanic["Pclass"].value_counts())

print("\nGender Distribution")
print(titanic["Sex"].value_counts())

plt.figure(figsize=(6,4))
plt.hist(titanic["Age"].dropna(), bins=20)
plt.title("Age Distribution")
plt.show()

titanic["Survived"].value_counts().plot(kind="bar")
plt.title("Survival Count")
plt.show()

print("\nCorrelation")
print(titanic.corr(numeric_only=True))

print("\nPost Lab Completed Successfully")
```

# Conclusion

Successfully performed Exploratory Data Analysis (EDA) on the Titanic dataset.

The dataset was analyzed using descriptive statistics, missing value analysis, and visualizations. Passenger survival, gender distribution, passenger class distribution, and age distribution were explored to understand the characteristics of the dataset.
